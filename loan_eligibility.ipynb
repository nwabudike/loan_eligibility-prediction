## Importing some essential libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

```

## importing dataset

```python
df=pd.read_csv('loan_data_set.csv')
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

## taking care of missing values


```python
df[['Dependents','Gender','Married','Self_Employed']]=df[['Dependents','Gender','Married','Self_Employed']].fillna(method='ffill')
```

```python
df[['LoanAmount','Loan_Amount_Term','Credit_History']]=df[['LoanAmount','Loan_Amount_Term','Credit_History']].fillna(df[['Dependents','LoanAmount','Loan_Amount_Term','Credit_History']].mean())
```

```python
df.isnull().sum()
```

## Encoding categorical columns using one hot encoding and label encoding

applying label encoding to gender,married,education,self_employed,property_area and loan_status

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df.Gender=le.fit_transform(df.Gender)
df.Married=le.fit_transform(df.Married)
df.Education=le.fit_transform(df.Education)
df.Self_Employed=le.fit_transform(df.Self_Employed)
df.Property_Area=le.fit_transform(df.Property_Area)
df.Loan_Status=le.fit_transform(df.Loan_Status)
df.Dependents=le.fit_transform(df.Dependents)
```

```python
df
```

## split our dataset into dependent nd independent variable and get our test nd train sets

```python
x=df.drop(['Loan_Status','Loan_ID'],axis=1).values
y=df.Loan_Status.values
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_train
```

Now we do some feature selection so as to reduce the dimensionality of this dataset,


## Performming  dimensionality Reduction(feature_selection)using LDA

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda=LDA(n_components=1)
x_train=lda.fit_transform(x_train,y_train)
x_test=lda.transform(x_test)
```

```python
x_train
```

## trainning our models with these dataset using kfold validation

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LogisticRegression(),x_train,y_train,cv=5)
```

```python
cross_val_score(SVC(kernel='rbf'),x_train,y_train,cv=5)
```

```python
cross_val_score(DecisionTreeClassifier(),x_train,y_train,cv=5)
```

```python
cross_val_score(RandomForestClassifier(n_estimators=100),x_train,y_train,cv=5)
```

from this you can see that LogisticRegression  is the best 

# our model


```python
model=SVC(kernel='rbf')
model.fit(x_train,y_train)
```

## prediction

```python
y_pred=model.predict(x_test)
y_pred
```

### model evaluation

```python
from sklearn.metrics import confusion_matrix , accuracy_score
cm=confusion_matrix(y_pred,y_test)
score=accuracy_score(y_pred,y_test)
cm
```

```python
score
```
