## Importing some essential libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

```

## importing dataset

```python
df=pd.read_csv('loan_data_set.csv')
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

notice we can remove the columns id,Date and zipcode because they don't add relevant impact to our model

```python
df=df.drop(['id','zipcode','date'],axis=1)
df.head()
```

get our dependent and independent variable

```python
x=df.drop('price',axis=1)
y=df.price
```

```python
x
```

Now we do some feature selection so as to reduce the dimensionality of this dataset,


## Performming  dimensionality Reduction(feature_selection)using backward elimination

```python
import statsmodels.api as sm
#first add a constant term to df_x
t=sm.add_constant(x)
dm=sm.OLS(y,t).fit()
dm.summary()
```

p_value of floor is high, we need to drop it


```python
x=x.drop('floors',axis=1)
```

```python
t=x
```

```python
dm=sm.OLS(y,t).fit()
dm.summary()
```

also p_value for sqft_lot is high and we need to drop it

```python
x=x.drop(['sqft_lot'],axis=1)
```

```python
t=x
```

```python
dm=sm.OLS(y,t).fit()
dm.summary()
```

now this is better

```python
x
```

## preformming split to the dataset to have test set and train set

```python
x=x.values
y=y.values
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

# performming feature scaling using standarzation

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_x_train=scaler.fit_transform(x_train)
scaled_x_test=scaler.transform(x_test)
```

## trainning our models with these dataset using kfold validation

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(LinearRegression(),scaled_x_train,y_train,cv=5)
```

```python
cross_val_score(SVR(),x_train,y_train,cv=5)
```

```python
cross_val_score(DecisionTreeRegressor(),scaled_x_train,y_train,cv=5)
```

```python
cross_val_score(RandomForestRegressor(),scaled_x_train,y_train,cv=5)
```

from this you can see that randomforestregressor is the best 

# our model


```python
model=RandomForestRegressor()
model.fit(scaled_x_train,y_train)
```

## prediction

```python
y_pred=model.predict(scaled_x_test)
y_pred
```

### model evaluation

```python
from sklearn.metrics import r2_score
r=r2_score(y_test,y_pred)
r
```

### save model

```python
import pickle
pickle.dump(model,open('Model_Predictor','wb'))
```
